# Symmetry

Symmetry is defined as invariance under a set of transformation and therefore, one defines a group as a collection of transformation. 

The branch of mathematics that deals with symmetry is called Group Theory. A group is a mathematical structure based on a set G with elements $\{a,b,c,\dots\}$, together with an operation $\cdot$, satisfying the following properties:

1. Closedness

    $\forall a,b \in G: a\cdot b \in G$

2. Associativity

    $\forall a,b,c \in G: (a\cdot y)\cdot z = a\cdot(y \cdot z)$

3. Identity element
 
    $\exists e \in G : \forall g\in G : e\cdot g = g\cdot e = g $
    
4. Inverses

    $\forall g\in G : \exists g^{-1} \in G :  g \cdot g^{-1} = g^{-1}\cdot g = e $
    
In addition, if $\forall a,b \in G : a\cdot b = b \cdot a $, then we say G is abelian (or commutative) group.

In quantum chemistry, for a target state $|\Psi \rangle$, the disired global symmetry $(S_G,N_G,I_G)$ can be imposed with $U(1) \otimes SU(2) \otimes P$. Where U(1) electron-number symmetry, SU(2) spin symmetry, and P the abelian point group symmetry with real-valued character tables.

\begin{equation}
P\in\{C_1,C_i,C_2,C_s,D_2,C_{2v},C_{2h},D_{2h}\}
\end{equation}

## Particle Number Symmetry U(1)

The global U(1) symmetry in Quantum Mechanics corresponds to the freedom to shift the phase of the wave function $|\Psi\rangle \rightarrow e^{iG} |\Psi^{\prime}\rangle$ and can be used to understand the conservation of electric charge.

G is the generator of the U(1) transformation, and G is a conserved observable if
the Hamiltonian H commutes with the U(1) transformation. Therefore, the energy
eigenstate can also be an eigenstate for the generator G:


The eigenvalue of G can be identified as the charge of a particle, for example.
Charge conservation is therefore a consequence of U(1) symmetry.

\begin{equation}
\hat{N} |\Psi\rangle= N|\Psi\rangle
\end{equation}

## Point Group Symmetry P


Point group symmetry is an important property of molecules widely used in some branches of chemistry: spectroscopy, quantum chemistry and crystallography. An individual point group is represented by a set of symmetry operations:

\begin{equation}
\begin{aligned}
L^2 |\Psi\rangle &= L(L+1)|\Psi\rangle\\
L_z |\Psi\rangle &= M_L|\Psi\rangle
\end{aligned}
\end{equation}

For diatomic and other linear molecules, only $L_z$ operation symmetry is satisfied, whereas, for nonlinear molecules, the wave function is an eigenfunction of the discrete symmetry operations R of the finite molecular point group.

### Symmetry Operations and Elements

The symmetry of a molecule consists of symmetry operations and symmetry elements. A symmetry operation is an operation that is performed to a molecule which leaves it indistinguishable and superimposable on the original position. Symmetry operations are performed with respect to symmetry elements (points, lines, or planes).

There are five types of symmetry operations including identity, reflection, inversion, proper rotation, and improper rotation. The improper rotation is the sum of a rotation followed by a reflection. The symmetry elements that correspond to the five types of symmetry operations are listed in Table.

### Abelian Point Group

Abelian groups are special types of groups in which commutativity holds. In other words, the binary operation on such groups is commutative. Because these abelian groups P all have real-valued character tables, the direct product of any irrep $I_j$ with itself gives the trivial irrep $I_0$:

\begin{equation}
\forall I_j: I_j\otimes I_j  = I_0
\end{equation}
